# 2D Channel Model
This is the basic setup that I am using to test the new LAYERS_THERMODYNAMICS features. It represents a 2D section view of the Southern Ocean. For now the equation of state is linear and there is no salinity. The surface forcing is just relaxation to a prescibed SST (timescale 30 days). There is also a sponge layer at the northern boundary with relaxation to a climatological temperature profile.

```python
%pylab inline
import mycolors
from mycolors import wocecolors
from MITgcmdata import MITgcmmodel
from seawater import pden

basedir = '/Users/rpa/RND/GCM/chan_2d/'
ddir = basedir + 'run_simple_fast'

m = MITgcmmodel.ModelInstance(ddir, default_iter=9473760, grid_type='sphericalpolar')
# coordinates for plotting
Zpc, Ypc = m.zf, hstack([m.yg, m.yg[-1]+2*(m.yc[-1]-m.yg[-1])])
# for contours
Z, Y = m.zc, m.yc.squeeze()
```

## Heat Budget
Here I show that I can close the heat budget at each grid point using the surface forcing, advective fluxes, and diffusive fluxes (implicit and explicit). The temperature equation is

$$ \frac{\partial \theta}{\partial t} = G_{adv} + G_{diff} + G_{forc} \ .$$

In the diffusive flux, both implicit and explicit contributions must be accounted for. The sponge layer is not included in this budget, but its effects are obvious at the northern boundary in the residual of the advective and diffusive fluxes.

```python
# this is what is in my diagnostic file
# 'TOTTTEND' 'ADVr_TH ' 'ADVx_TH ' 'ADVy_TH ' 'DFrE_TH ' 'DFxE_TH ' 'DFyE_TH ' 'DFrI_TH '

d = m.rdmds('DiagGAD-T', useMask=False).squeeze()

tottend = d[0]
ADVr_TH, ADVx_TH, ADVy_TH = d[1], d[2], d[3]
DFrE_TH, DFxE_TH, DFyE_TH, DFrI_TH = d[4],d[5],d[6],d[7]
vol = ma.masked_equal( m.hFacC.squeeze() * m.rac.squeeze()[newaxis,:] * m.dzf[:,newaxis], 0)
tend_adv = -vol**-1*( (ADVr_TH - vstack([ADVr_TH[1:], zeros((1,m.Ny))]))
            +(roll(ADVy_TH,-1,axis=1) - ADVy_TH) )
tend_diff_rE = -vol**-1*( (DFrE_TH - vstack([DFrE_TH[1:], zeros((1,m.Ny))])) )
tend_diff_rI = -vol**-1*( (DFrI_TH - vstack([DFrI_TH[1:], zeros((1,m.Ny))])) )
tend_diff_yE = -vol**-1*( (roll(DFyE_TH,-1,axis=1) - DFyE_TH) )
tend_diff = tend_diff_yE + tend_diff_rE + tend_diff_rI
tottend = ma.masked_array(tottend, vol.mask)

# tendency from forcing
rhonil, HeatCapacity_Cp = 1035., 3994.
# remember TFLUX is positive down (different from adv and diff fluxes)
tend_forc = vstack([ ma.masked_invalid( Tflux /
            (m.hFacC[0,:,0]*m.dzf[0]) / rhonil / HeatCapacity_Cp ),
            ma.masked_array( zeros((m.Nz-1,m.Ny)), vol[1:].mask ) ])

zmax = -5000
day = 24*60*60

figure(figsize=(16,10))

subplot(221, axisbg='k')
pcolormesh(Ypc, Zpc, tend_adv*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.001,0.001]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('Tendency from Advection')

subplot(222, axisbg='k')
pcolormesh(Ypc, Zpc, tend_diff*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.001,0.001]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('Tendency from Diffusion')

subplot(223, axisbg='k')
pcolormesh(Ypc, Zpc, (tend_diff+tend_adv)*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.0003,0.0003]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('Sum of Advection and Diffusion')

subplot(224, axisbg='k')
pcolormesh(Ypc, Zpc, tottend, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.0003,0.0003]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('TOTTEND')


figure(figsize=(16,10))
zmax = -300

subplot(321, axisbg='k')
pcolormesh(Ypc, Zpc, tend_adv*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.01,0.01]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
ylabel('depth (m)')
title('Tendency from Advection')

subplot(322, axisbg='k')
pcolormesh(Ypc, Zpc, tend_diff*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.01,0.01]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
ylabel('depth (m)')
title('Tendency from Diffusion')


subplot(323, axisbg='k')
pcolormesh(Ypc, Zpc, (tend_diff+tend_adv)*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.01,0.01]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
ylabel('depth (m)')
title('Sum of Advection and Diffusion')

subplot(324, axisbg='k')
pcolormesh(Ypc, Zpc, tend_forc*day, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.01,0.01]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
ylabel('depth (m)')
title('Surface Forcing')

subplot(325, axisbg='k')
pcolormesh(Ypc, Zpc, (tend_diff+tend_adv+tend_forc)*day,
           cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.0003,0.0003]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('Surface Forcing + Advection + Diffusion')

subplot(326, axisbg='k')
pcolormesh(Ypc, Zpc, tottend, cmap=get_cmap('RdBu_r')); colorbar(); 
clim([-0.0003,0.0003]);
contour(Y,Z, T, 12, colors='k') 
ylim([zmax,0])
xlabel('lat'); ylabel('depth (m)')
title('TOTTEND')

```

# LAYERS Output: Isopycnal and Diapycnal Volume Fluxes

### Water Mass Transformation Equations

In isopycnal coordinates (using $\theta$ as our density variable), the mass conservation equation takes the form of a thickness equation

$$ \partial_t \sigma + \partial_x (u \sigma) + \partial_y (v \sigma) + \partial_\theta (G \sigma) = 0 $$

where $\sigma = (\partial \theta / \partial z)^{-1}$ is the continuous layer thickness and $G$ represents all _non-advective_ tendencies from the RHS of the $\theta$ equation. From now on we will discard the $x$-dependent $u$ term, since our domain is 2D ($y,z$).

To derive the water mass transformation equation, we integrate the thickness equation from the sea floor up to an arbitrary isopycnal $\theta$, zonally in $x$, and from the southern boundary up to an arbitrary latitude $y$. For example, applying this integral directly to the thickness $\sigma$ yeilds

$$ \oint_x\int_{y_{min}}^y  \int_{bot}^\theta \sigma(y',\theta') d\theta' dy' dx =
   \oint_x \int_{y_{min}}^y \int_{-H}^0 \mathcal{H}(-\theta) dz dy' dx \equiv V(y,\theta) $$

where we have used the equality $\sigma d\theta' = dz$ along with the [Heaviside function](http://en.wikipedia.org/wiki/Heaviside_step_function) $\mathcal{H}$ to convert the integral in $\theta'$ to and integral in $z$. The physical interpretation of $V(y,\theta)$ is _the volume enclosed by the isopycnal $\theta$ and the latitude circle $y$_. Applying the same integration to each of the terms in the thickness equation gives

$$ \partial_t V(y,\theta) + \Psi(y,\theta) + \Omega(y,\theta) = 0 \ . $$

The volume flux term $\Psi$ is defined as

$$ \Psi(y,\theta) \equiv \oint_x \int_{y_{min}}^y \int_{bot}^\theta \partial_y( v \sigma) d\theta' dy' dx
 = \oint_x \int_{bot}^\theta -(v \sigma)\big |_y d\theta' dy' dx
 = -\oint_x \int_{-H}^0 v \mathcal{H}(-\theta) dz dx \ .$$

where we have assumed there is no flow through the southern boundary. The transformation $\Omega$ is defined as

$$ \Omega(y,\theta) \equiv \oint_x  \int_{y_{min}}^y \int_{bot}^\theta \partial_\theta(\sigma G) d\theta' dy' dx
 = \oint_x  \int_{y_{min}}^y (G \sigma) \big |_\theta dy' dx
 = \partial_\theta \oint_x \int_{y_{min}}^y \int_{-H}^0 G \mathcal{H}(-\theta) dz dx $$

### LAYERS Diagnostics

The LAYERS package makes it possible to compute a discrete version of the transformation equation. The isopycnal layers boundaries are defined by the user. If $N_l$ is the number of layers, there are $N_l + 1$ boundaries. $\theta_{k+\frac{1}{2}}$ denotes the upper boundary of layer $k$.

LAYERS computes the time-averaged volume flux in each discrete layer as

$$ (\overline{vh})_{j,k} = \frac{1}{\Delta t} \int_t^{t+\Delta t} \int_{-H}^0 v \mathcal{H}(\theta_{k-\frac{1}{2}}) \mathcal{H}(-\theta_{k+\frac{1}{2}})  dz dt \ . $$

This makes it possible to construct $\Psi$ by summing in layers, i.e.

$$ \Psi_{j,k+\frac{1}{2}} = \sum\limits_{k'=0}^k \Delta x_g (\overline{vh})_{j,k'} $$

where the subscript $k+\frac{1}{2}$ indicates that $\Psi$ is defined at the upper boundary of the layer.

For $\Omega$ it is a bit more complicated, because we have to take a derivative with respect to $\theta$. LAYERS keeps track of the diapycnal terms in the same way, i.e.

$$ (\overline{Gh})_{j,k} = \frac{1}{\Delta t} \int_t^{t+\Delta t} \int_{-H}^0 G \mathcal{H}(\theta_{k-\frac{1}{2}}) \mathcal{H}(-\theta_{k+\frac{1}{2}})  dz dt $$

and does so separately for the surface forcing, vertical mixing, and horizontal mixing. We define an intermediate function $F$ as the integral in $y$ and $\theta$:

$$ F_{j,k+\frac{1}{2}} = \sum\limits_{j'=0}^{j-1} \sum\limits_{k'=0}^k A_c (\overline{Gh})_{j',k'} $$

To directly compare $\Psi$ and $\Omega$, we must keep $\Omega$ on the $k+\frac{1}{2}$ point by taking a second-order centered difference-derivative:

$$ \Omega_{j,k+\frac{1}{2}} = \frac{ F_{j,k+\frac{3}{2}} - F_{j,k-\frac{1}{2}} }{ \theta_{k+\frac{3}{2}} - \theta_{k-\frac{1}{2}} } $$

Below we calculate $\Omega_{j,k+\frac{1}{2}}$ and $\Psi_{j,k+\frac{1}{2}}$. We contour them in the $y,\theta$ plane and plot them at two locations: just below the base of the mixed layer, and at the northern boundary just in front of the sponge layer.

```python
# LAYERS fields:
#'LaVH1TH ' 'LaHs1TH ' 'LaPs1TH ' 'LaHc1TH ' 'LaPc1TH ' 'LaTs1TH ' 'LaTh1TH ' 'LaTr1TH '
dl = m.rdmds('DiagLAYERS', useMask=False)

from MITgcmutils.mds import rdmds
Thl = rdmds(m.output_dir + '/layers1TH').squeeze()
#Thl = Tlevs.copy()
Thc = 0.5*(Thl[1:] + Thl[:-1])
# volume
Vol = dl[1].squeeze() * m.rac.squeeze()[newaxis,:]
Nl, Ny = len(Thc), m.Ny
# transport
VH = dl[0].squeeze() * m.dxg.squeeze()[newaxis,:]
# streamfunction, defined on the v, k+1/2 points
Psi = -cumsum(VH,axis=0)

# diapycnal term
GH = dl[5:].squeeze() * m.rac[newaxis,newaxis,:,0]
# formation rate
F = GH.cumsum(axis=2).cumsum(axis=1)
# F is defined as the sum up to but excluding the current j index
F = dstack( [ zeros((3,Nl,1)), F[:,:,:-1] ] )
# fill F in with a zero row
F = hstack( [ zeros((3,1,Ny)), F ] )
Om = (F[:,2:] - F[:,:-2]) / (Thl[2:] - Thl[:-2])[newaxis,:,newaxis]
```

```python
Thl_scale = log2(Thl - Thl.min()+1)
Thc_scale = log2(Thc - Thl.min()+1)
#Thl_scale = hstack([0, Vol.sum(axis=1).cumsum(axis=0)])
#Thc_scale = 0.5*(Thl_scale[1:] + Thl_scale[:-1])
Thl_lab = ['%4.2f' % lab for lab in Thl ]

scalefac = 720./1e6
psilevs = arange(-32,32,2)+1
psiticks = arange(-30,31,10)


figure(figsize=(16,8))
subplot(121)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*Om.sum(axis=0), psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Omega(y,\theta)$ Total (Sv)')
subplot(122)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*Om[0], psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Omega(y,\theta)$ Surface (Sv)')

figure(figsize=(16,8))
subplot(121)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*Om[1], psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Omega(y,\theta)$ Lateral Mixing (Sv)')
subplot(122)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*Om[2], psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Omega(y,\theta)$ Vertical Mixing (Sv)')


figure(figsize=(16,8))
subplot(121)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*Psi[:-1], psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Psi(y,\theta)$ (Sv)')
subplot(122)
contourf( Ypc[:-1], Thl_scale[1:-1], 
         scalefac*(Psi[:-1] - Om.sum(axis=0)), psilevs,
         extend='both', cmap=get_cmap('RdBu_r'));
yticks(Thl_scale[1:-1],Thl_lab[1:-1]); grid()
colorbar(orientation='horizontal', ticks=psiticks, shrink=0.6)
xlabel('lat'); ylabel(r'$\theta$')
title(r'$\Psi - \Omega$ (Sv)')
```

# Conclusions

Overall the agreement between $\Psi$ and $\Omega$ looks pretty good. The fact that it is not perfect can be attributed to three different possibilities:

1. The model is not in a completely steady state (i.e. there are trends in $V$). This is unlikely because it has been run for 730 years and all monintor statistcs are stable.

1. There is unresolved numerical diffusion. This seems very likely and could probably be assessed by changing advection schemes.

1. The LAYERS numerical scheme is imperfect. I suspect this is also true, but I don't know how to assess it.
