```python
from mitgcmdata import MITgcmmodel
import MITgcmutils.mds as mds
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import SymLogNorm
%matplotlib inline
```

```python
ddir = '/glade/scratch/rpa/SOSE/run_layers'
m = MITgcmmodel.ModelInstance(ddir, default_iter=105168)
```

```python
lb = mds.rdmds(ddir + '/layers1RHO')
dlt = m.rdmds('DiagLAYERS-transport')
# field_list = 'LaUH1RHO' 'LaHw1RHO' 'LaPw1RHO' 'LaVH1RHO' 'LaHs1RHO' 'LaPs1RHO'
dld = m.rdmds('DiagLAYERS-diapycnal')
# fldList = { 'LaTs1RHO' 'LaTh1RHO' 'LaTr1RHO' 'LaSs1RHO' 'LaSh1RHO' 'LaSr1RHO' 'LaHc1RHO' 'LaPc1RHO' }
```

```python
# calculate streamfunction
Pis = dlt[5]

# Pi, a measure of outcropping
# (not so easy to interpret with complex topography)
lat_circ = (~m.mask[0]).sum(axis=1)[np.newaxis,]
Pis_scaled  = 1 - (lat_circ - dld[7].sum(axis=2))/m.Nx
# a simpler measure
surface_mask = (dld[0]!=0.).astype('f8')

VH = dlt[3]
Psi = (VH * m.dxg).sum(axis=2).cumsum(axis=0)
```

```python
Nl = len(lb)-1

fig = plt.figure(figsize=(18,10))
Psi_levs = np.arange(-30,30,2) + 1
ax = fig.add_subplot(111)
cf = ax.contourf(m.yg[0,:,0].squeeze(), np.arange(Nl), Psi/1e6,
            Psi_levs, cmap='RdBu_r', extend='both')
ax.contour(m.yg[0,:,0].squeeze(), np.arange(Nl)+0.5, surface_mask.mean(axis=2), [0.1, 0.5, 0.9], colors='0.3')
plt.ylim([Nl-10,4])
plt.xlim([m.yg[0,0,0], -30])
ax.set_yticks(np.arange(0,Nl,2))
rho_labels = ['%4.2f' % r for r in lb[np.arange(0,Nl,2)].squeeze()]
ax.set_yticklabels(rho_labels)
plt.colorbar(cf, ticks=range(-40,41,5))
ax.grid()
ax.set_title('Isopycnal Streamfunction')
ax.set_ylabel(r'$\sigma_0$')
ax.set_xlabel('latitude')
```

```python
for j in [m.Ny-33, m.Ny-33-120]:

    X = np.hstack([m.xg[0,j], m.xg[0,j,-1]+(1/6.)])

    fig = plt.figure(figsize=(18,10))
    ax = plt.subplot2grid((1,5),(0, 1), colspan=4)
    pc = ax.pcolormesh(X, np.arange(Nl+1), VH[:,j,:], cmap='RdBu_r',
                    norm=SymLogNorm(linthresh=1, vmin=-120, vmax=120))
    #plt.clim([-100,100])
    #plt.colorbar(pc, shrink=0.5)
    ax.set_ylim([Nl,1])
    ax.set_xlim([0,360])
    ax.set_yticks(np.arange(0,Nl,2))
    rho_labels = ['%4.2f' % r for r in lb[np.arange(0,Nl,2)].squeeze()]
    ax.set_yticklabels([])
    ax.set_xticks(range(0,360,30))
    ax.set_ylabel(r'$\sigma_0$')
    ax.set_xlabel('longitude')
    ax.grid()
    ax.set_title(r'Meridional Isopycnal Transport (m$^2$ s$^{-1}$) at %2d S' % (-m.yc[0,j,0]))

    ax2 = plt.subplot2grid((1,5),(0, 0), axisbg='none')
    ax2.plot(Psi[:,j]/1e6, np.arange(Nl)+1, 'k.-')
    ax2.set_ylim([Nl,1])
    ax2.set_yticks(np.arange(0,Nl,2))
    ax2.set_yticklabels(rho_labels)

    ax2.set_xlabel('Sv')
    ax2.set_title('Isopycnal Streamfunction')
    ax2.grid()
```

```python
X = np.hstack([m.xg[0,j], m.xg[0,j,-1]+(1/6.)])
Y = np.hstack([m.yg[0,:,0], m.yg[0,-1,0]+(1/6.)])
L = np.arange(Nl+1)

# dld fields
# 'LaTs1RHO' 'LaTh1RHO' 'LaTr1RHO' 'LaSs1RHO' 'LaSh1RHO' 'LaSr1RHO' 'LaHc1RHO' 'LaPc1RHO' 

field_idx = [0,3,1,4,2,5]
field_name = ['Surface Heat Flux', 'Surface Salt Flux',
              'Lateral Mixing of Temperature', 'Lateral Mixing of Salt',
              'Vertical Mixing of Temperature', 'Vertical Mixing of Salt']

rhoticks = np.arange(0,Nl,2)
rho_labels = ['%4.2f' % r for r in lb[rhoticks].squeeze()]


fig = plt.figure(figsize=(18,20))
for n, f in enumerate(zip(field_idx, field_name)):


    ax = fig.add_subplot(3,2, n+1)
    pc = ax.pcolormesh(Y, L, dld[f[0]].sum(axis=2), cmap='RdBu_r')
    pc.set_clim([-1e-4,1e-4])
    cf = ax.contour(m.yg[0,:,0].squeeze(), np.arange(Nl), Psi/1e6,
            Psi_levs[::2], colors='k')
    #ax.contour(m.yg[0,:,0].squeeze(), np.arange(Nl)+0.5, surface_mask.mean(axis=2), [0.1, 0.5, 0.9], colors='0.6')
    plt.ylim([Nl-10,4])
    plt.xlim([m.yg[0,0,0], -30])
    ax.set_yticks(rhoticks)
    #ax.set_yticklabels(rho_labels);
    ax.set_yticklabels([]);
    plt.title(f[1])
```

```python
#fig = plt.figure(figsize=(12,6))
j30 = np.argmin((m.yc[0,:,0]+30)**2)
drho = np.diff(lb, axis=0)
Sv=1e6

# net transformation rate
T = (dld[:6] * m.rac[np.newaxis,:,:,:] / drho[np.newaxis,:,:,:])[:,:,:j30].sum(axis=3).sum(axis=2).T
```

```python
plt.figure(figsize=(12,6))
plt.plot(lb[:-1].squeeze() - 1000., T.sum(axis=1)/Sv, 'k', linewidth=2)
plt.plot(lb[:-1].squeeze() - 1000., T[:,field_idx]/Sv)
plt.xlim([23,28.5])
plt.xlabel(r'$\sigma_0$')
plt.ylabel('Sv')
plt.legend(['Net'] + field_name, loc='upper left')
```
