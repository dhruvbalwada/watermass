```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib
import brewer2mpl
from mitgcmdata import MITgcmmodel
from jmd95 import eos

%matplotlib inline
plt.rcParams['image.origin'] = 'bottom'
plt.rcParams['figure.figsize'] = (16,8)

bmap = brewer2mpl.get_map('Set1', 'Qualitative', 6)
plt.rcParams['axes.color_cycle'] = bmap.mpl_colors
```

```python
ddir = '/glade/scratch/rpa/SOSE/run_matt'

m = MITgcmmodel.ModelInstance(output_dir=ddir)
```

```python
# monthly stuff
deltaT = 900
#diag_freq = 2629200. # monthly
#diag_prefix = 'DiagIce-monthly'

diag_freq = 432000 # 5-day
diag_prefix= 'DiagIce-5day'

diag_delta_iter = diag_freq/deltaT

ntimesteps = 210336
diag_iters = np.round(np.arange(diag_delta_iter, ntimesteps+1, diag_delta_iter)).astype('i4')

diag_names = ['SIarea', 'SIheff', 'SIhsnow', 'SItflux', 'SIatmQnt', 'SIempmr',
             'SIatmFW', 'SIuheff', 'SIvheff', 'TFLUX', 'SFLUX', 'THETA', 'SALT']
```

```python
# connect to cluster
from IPython.parallel import Client, interactive
c = Client(profile='default')
dview = c.direct_view()
lbview = c.load_balanced_view()
```

```python
with dview.sync_imports():
    import numpy
    from MITgcmutils import mds, wrmds
    from pygamman import gamman
```

```python
# push info to nodes
dview.push(dict(
    diag_theta_fname = m.output_dir + '/' + diag_prefix,
    diag_salt_fname = m.output_dir + '/' + diag_prefix,
    diag_theta_rec = 11,
    diag_salt_rec = 12,
    output_dir = m.output_dir + '/gamma_n_surface/',
    lon = m.dxc[0],
    lat = m.dyc[0]
))
```

```python
# function for neutral density calc
# it should be as simple as possible
@interactive
def calc_gamma_n(niter):
    theta = mds.rdmds(diag_theta_fname, niter, rec=diag_theta_rec)
    salt = mds.rdmds(diag_salt_fname, niter, rec=diag_salt_rec)
    pres = numpy.zeros_like(theta)
    N = theta.size
    
    mask = (salt==0.)
    [s, t, p, lo, la] = [numpy.ma.masked_array(f, mask).filled(0.).ravel()
                         for f in (salt, theta, pres, lon ,lat)]
    
    gamma, dglo, dghi = gamman.gamma_n(s, t, p, N, lo, la)
    gamma.shape = lon.shape
    
    wrmds(output_dir + 'gamma_n_surface', gamma, niter)

```

```python
res = lbview.map_async(calc_gamma_n, diag_iters)
```
