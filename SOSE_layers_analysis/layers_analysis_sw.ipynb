```python
def process_layers(ddir, iter0, diters, N, deltaTclock=900, layers_name='1RHO'):

    iters = range(iter0,iter0+N*diters,diters)
    DT = diters * deltaTclock
    m = MITgcmmodel.ModelInstance(ddir, default_iter=list(iters))


    layers_transport = m.rdmds('DiagLAYERS-transport', iters[1:]).mean(axis=0)
    uh, vh = layers_transport[0], layers_transport[3]
    hw, hs = layers_transport[1,::-1], layers_transport[4,::-1]
    pi_s = layers_transport[5,::-1]
    
    layers_diapycnal = m.rdmds('DiagLAYERS-diapycnal', iters[1:]).mean(axis=0)
    layers_trend = m.rdmds('DiagLAYERS-trend', iters)

    # process data
    la = layers.LayersAnalyzer(m, layers_name=layers_name)
    dvol_dt = la.volume_trend(layers_trend[np.r_[0,-1]], DT*(N-1))
    #dvol_dt = la.volume_trend(layers_trend, DT)
    #if dvol_dt.ndim == 4:
    #    # average over time
    #    dvol_dt = dvol_dt.mean(axis=0)

    # everything but the advective tendencies
    diapycnal_vel = la.diapycnal_velocity(layers_diapycnal[np.r_[:3,5:8]])

    advective_vel = la.advective_flux_divergence(uh, vh)
    
    # "direct" form of advective vel
    advective_vel_direct = la.diapycnal_velocity(layers_diapycnal[np.r_[3,4,8,9]]).sum(axis=0)
    
    # "TOTTEND"
    diapycnal_vel_tottend = -la.diapycnal_velocity(layers_diapycnal[np.r_[10,11]]/(24*60*60)).sum(axis=0)
    
    diapycnal_vel_numerical = dvol_dt - diapycnal_vel.sum(axis=0) - advective_vel
    diapycnal_vel_numerical_direct = dvol_dt - diapycnal_vel.sum(axis=0) - advective_vel_direct

    
    psi = la.calculate_moc(vh) / 1e6
    print psi.shape

    # something is wrong with this
    pi = pi_s.mean(axis=-1)
    
    return {'dvol_dt': dvol_dt,
            'diapycnal_vel': diapycnal_vel,
            'advective_vel': advective_vel,
            'advective_vel_direct': advective_vel_direct,
            'diapycnal_vel_numerical': diapycnal_vel_numerical,
            'diapycnal_vel_tottend': diapycnal_vel_tottend,
            'diapycnal_vel_numerical_direct': diapycnal_vel_numerical_direct,
            'uh': uh, 'vh': vh, 'hs': hs, 'hw': hw,
            'psi': psi, 'pi': pi, 'la': la, 'm': m}

def plot_diapycnal_velocities(d, r=0.7e6, layers_lim=[28.2, 22], depth_lim=[-5000,0],
                              direct=False, zcoord=True):
    
    if direct:
        diapyc_suff = '_direct'
    else:
        diapyc_suff = ''

    m = d['m']
    Yg = np.hstack([m.yg[0,:,0], 2*m.yg[0,-1,0]-m.yg[0,-2,0]])
    Xg = np.hstack([m.xg[0,0,:], 2*m.xg[0,0,-1]-m.xg[0,0,-2]])
    Yc = m.yc[0,:,0]
    Xc = m.xc[0,0,:]

    lb = d['la'].layers_bounds_w - 1000.
    lbc = d['la'].layers_top - 1000.

    if zcoord:
        zw = np.cumsum(d['hs'].mean(axis=-1)[::-1], axis=0)
        # index goes down from the surface
        # pcolor needs boundaries
        zc = -np.vstack([zw[0][np.newaxis,:]/2,
                        0.5*(zw[1:] + zw[:-1])])[::-1]
        zc = np.hstack([zc, zc[:,-1][:,np.newaxis]])
        Ygz = np.tile(Yg[np.newaxis,:], (zc.shape[0],1))

        zw = -np.vstack([np.zeros((1,m.Ny)), zw[:-1]])[::-1]
        Ycz = np.tile(Yc[np.newaxis,:], (zw.shape[0],1))
        ylim=[-5000,0]

        Xpc = Ygz
        Ypc = zc
        Xc = Ycz
        Yc = zw
        ylim = depth_lim
    else:
        Xpc = Yg
        Ypc = lb
        Xc = Yc
        Yc = lbc
        ylim = layers_lim
        
    
    plt.figure(figsize=(16,12))

    plt.subplot(221)
    plt.pcolormesh(Xpc, Ypc, d['diapycnal_vel'].sum(axis=0).sum(axis=-1), cmap='RdBu_r')
    plt.clim([-r, r])
    plt.colorbar()
    plt.contour(Xc, Yc, d['psi'], 10, colors='k')
    #plt.contour(Yg[:-1], lbc, pi, [0.05, 0.5, 0.95], colors='y')
    plt.ylim(ylim)
    plt.title('Explicit Diapycnal Velocity')

    plt.subplot(222)
    plt.pcolormesh(Xpc, Ypc, d['dvol_dt'].sum(axis=-1), cmap='RdBu_r')
    plt.clim([-r, r])
    plt.colorbar()
    plt.contour(Xc, Yc, d['psi'], 10, colors='k')
    plt.ylim(ylim)
    plt.title('Volume Trend')

    plt.subplot(223)
    plt.pcolormesh(Xpc, Ypc, d['advective_vel' + diapyc_suff].sum(axis=-1), cmap='RdBu_r')
    plt.clim([-r, r])
    plt.colorbar()
    plt.contour(Xc, Yc, d['psi'], 10, colors='k')
    plt.ylim(ylim)
    plt.title('Advective Diapycnal Velocity')

    plt.subplot(224)
    plt.pcolormesh(Xpc, Ypc, d['diapycnal_vel_numerical' + diapyc_suff].sum(axis=-1), cmap='RdBu_r')
    plt.clim([-r, r])
    plt.colorbar()
    plt.contour(Xc, Yc, d['psi'], 10, colors='k')
    plt.ylim(ylim)
    plt.title('Numerical Diapycnal Velocity')
    
def layers_plot(Xpc, Ypc, field, r, Xc, Yc, psi, ylim, title, sigma=None, ax=None):
    if ax is None:
        ax = plt.subplot(111)
    pc = ax.pcolormesh(Xpc, Ypc, field, cmap='RdBu_r')
    pc.set_clim([-r, r])
    plt.colorbar(pc)
    ax.contour(Xc, Yc, psi, 16, colors='0.5')
    ax.set_ylim(ylim)
    ax.set_title(title)
    if sigma is not None:
        c = ax.contour(Xc, Yc, sigma, [27.9,27.5,27.2,27.0,26.5,26.0], colors='k')
        zloc = -400
        manual_locations =[(-70, zloc), (-55, zloc), 
                           (-50, zloc), (-40, zloc), (-30, zloc)]
        plt.clabel(c, fmt='%1.1f', 
         manual=manual_locations)

        
```
