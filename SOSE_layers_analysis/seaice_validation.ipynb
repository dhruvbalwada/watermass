```python
# monthly stuff
deltaT = 900
diag_freq = 2629200.
diag_delta_iter = diag_freq/deltaT

ntimesteps = 210336
diag_iters = np.round(np.arange(diag_delta_iter, ntimesteps+1, diag_delta_iter)).astype('i4')
Nt = len(diag_iters)
# starts in January
months = np.tile(np.arange(1,13), Nt/12)

# season definitions
# easy because we want to start with january
season = np.tile(np.repeat(np.arange(4),3), Nt/12)
season_names = ['JFM', 'AMJ', 'JAS', 'OND']
season_idx = {}
for n,s in enumerate(season_names):
    season_idx[s] = list(diag_iters[season==n])
```

```python
diag_names = ['SIarea', 'SIheff', 'SIhsnow', 'SItflux', 'SIatmQnt', 'SIempmr',
             'SIatmFW', 'SIuheff', 'SIvheff', 'TFLUX', 'SFLUX', 'THETA', 'SALT']
diag_prefix = 'DiagIce-monthly'
data = dict((el,dict()) for el in diag_names)
```

```python
for s in season_names:
    iters = list(season_idx[s])
    d = m.rdmds(diag_prefix, iters).mean(axis=0)
    for n, dn in enumerate(diag_names):
        data[dn][s] = d[n]
        
```

```python
for dn in diag_names:
    data[dn]['ANN'] = 0.25*(data[dn]['JFM'] + data[dn]['AMJ']
                           +data[dn]['JAS'] + data[dn]['OND'])
#for dname = diag_names
```

```python
from mpl_toolkits.basemap import Basemap
def southern_ocean_pcolormesh(lon, lat, data, boundinglat=-55, lw=0.25,
                              ax=None, labels=[1,1,0,1], **kwargs):
    """Plot something in the southern ocean."""
    
    m = Basemap(projection='spstere',boundinglat=boundinglat,lon_0=180, ax=ax)
    #m.drawcoastlines()
    m.fillcontinents(color='0.5',lake_color='0.55')
    # draw parallels and meridians.
    m.drawparallels(np.arange(-80.,81.,20.), linewidth=lw)
    m.drawmeridians(np.arange(-180.,181.,30.), labels=labels, linewidth=lw)
    m.drawmapboundary(fill_color='w')
    #x, y = m(nc.variables['TLON'][:], nc.variables['TLAT'][:])
    return m.pcolormesh(lon, lat, data, latlon=True, **kwargs), m

def southern_ocean_contourf(lon, lat, data, clevs, boundinglat=-55, lw=0.25, 
                            ax=None, labels=[1,1,0,1], **kwargs):
    """Plot something in the southern ocean."""
    
    m = Basemap(projection='spstere',boundinglat=boundinglat,lon_0=180, ax=ax)
    #m.drawcoastlines()
    m.fillcontinents(color='0.5',lake_color='0.55')
    # draw parallels and meridians.
    m.drawparallels(np.arange(-80.,81.,20.), linewidth=lw)
    m.drawmeridians(np.arange(-180.,181.,30.), labels=labels, linewidth=lw)
    m.drawmapboundary(fill_color='w')
    #x, y = m(nc.variables['TLON'][:], nc.variables['TLAT'][:])
    return m.contourf(lon, lat, data, clevs, latlon=True, **kwargs), m

```

```python
lon = m.xc[0,:,:]
lat = m.yc[0,:,:]
```

# Ice Thickness #

Compare to Holland et al. (2014) Fig. 3.

```python
fig = plt.figure(figsize=(15,16))
clevs = np.arange(0,2.1,0.2)
for n, season in enumerate(season_names):
    mask = (data['SIarea'][season] < 0.5)
    fld = np.ma.masked_array(data['SIheff'][season], mask)
    
    ax = fig.add_subplot(2,2,n+1)
    pc, bm = southern_ocean_contourf(lon, lat, fld, clevs, ax=ax, extend='max', cmap='jet')#cmap='GnBu')
    #pc.set_clim([0,2])
    ax.set_title(season.upper())

cax = fig.add_axes([0.25,0.05,0.5,0.02])
cb = plt.colorbar(pc, cax=cax, orientation='horizontal')
cb.set_label('Ice Thickness')
plt.show()
```

# Ice Concentration and Velocity #

Compare to Holland et al. (2014) Fig. 5.

```python
fig = plt.figure(figsize=(15,16))
clevs = np.arange(0,1.1,0.1)

sx,sy = 12,6 # subsample 

for n, season in enumerate(season_names):
    area = np.ma.masked_less(data['SIarea'][season], 0.1)
    mask = area.mask
    H = np.ma.masked_array(data['SIheff'][season], mask)
    UH, VH = data['SIuheff'][season], data['SIvheff'][season]

    U = np.ma.masked_invalid(UH/H)
    V = np.ma.masked_invalid(VH/H)
    Urot, Vrot = bm.rotate_vector(U, V, lon, lat)
    
    ax = fig.add_subplot(2,2,n+1)
    pc, bm = southern_ocean_contourf(lon, lat, area, clevs, ax=ax, extend='max', cmap='jet')#cmap='GnBu')
    
    xx, yy = bm(lon, lat)
    q = bm.quiver(xx[::sy,::sx], yy[::sy,::sx], -Urot[::sy,::sx], -Vrot[::sy,::sx], scale=5)
    
    qk = plt.quiverkey(q, 0.1, 0.1, 0.1, label='0.1 m/s')
    
    ax.set_title(season.upper())

cax = fig.add_axes([0.25,0.05,0.5,0.02])
cb = plt.colorbar(pc, cax=cax, orientation='horizontal')
cb.set_label('Ice Concentration and Velocity')
plt.show()
```

# Freshwater Flux #

```python
fig = plt.figure(figsize=(15,16))
rholevs = [26.5, 27., 27.2, 27.6, 28.0, 28.2]
clevs = np.arange(-3,3,0.25)+0.125

# gives m/year
scalefac = 24*60*60*365/1e3

for n, season in enumerate(season_names):
    
    fld1 = scalefac * np.ma.masked_array(data['SIatmFW'][season], m.mask[0])
    fld2 = -scalefac * np.ma.masked_array(data['SIempmr'][season], m.mask[0])
    #ice_fw = fld2 - fld1
    ice_fw = fld2
    
    ax = fig.add_subplot(2,2,n+1)
    pc, bm = southern_ocean_contourf(lon, lat, ice_fw, clevs,
                                   ax=ax, extend='both', cmap='jet')# cmap='PiYG')
    
    rho, drhodt, drhods = eos.state_surface(data['THETA'][season], data['SALT'][season])
    con = bm.contour(lon, lat, np.ma.masked_array(rho-1000, m.mask[0]), rholevs,
               latlon=True, colors='k')
    plt.clabel(con, fmt='%3.1f')
    #pc.set_clim([-3,3])
    ax.set_title(season.upper())

cax = fig.add_axes([0.25,0.05,0.5,0.02])
cb = plt.colorbar(pc, ticks=np.arange(-2.5,2.6,0.5), cax=cax, orientation='horizontal')
cb.set_label('Freshwater Flux (m/year)')
plt.show()
```
