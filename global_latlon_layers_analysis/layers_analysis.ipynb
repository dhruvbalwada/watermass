```python
# The layers are indexed starting from lowest tracer value to highest
# for temperature, this means that the lowest indices correspond to 
# the deepest layers.
# But for density the order has to be reversed.
layers_rho = True
if layers_rho:
    ksign = -1
else:
    ksign = 1
    
lb = mds.rdmds(m.output_dir + '/layers1RHO')[::ksign]
drho = np.diff(lb,axis=0)
# center density
rhoc = 0.5*(lb[:-1,0,0] + lb[1:,0,0]) - 1000.
# density at the top of the column
rhot = lb[1:,0,0] - 1000.

# load data
UH, Hw, Piw, VH, Hs, Pis = m.rdmds('DiagLAYERS-transport')[:,::ksign]
Ts, Th, Tr, Tha, Tra, Ss, Sh, Sr, Sha, Sra = m.rdmds('DiagLAYERS-diapycnal')[:,::ksign]
Hs0,Hw0,Hc0 = m.rdmds('DiagLAYERS-trend', iter0)[:,::ksign]
Hs1,Hw1,Hc1 = m.rdmds('DiagLAYERS-trend', iter1)[:,::ksign]
```

```python
# transformation terms
Om_Ts = (Ts * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Th = (Th * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Tr = (Tr * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Ss = (Ss * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Sh = (Sh * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Sr = (Sr * m.rac).sum(axis=-1).cumsum(axis=-1)

# advective stuff
Om_Tha = (Tha * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Tra = (Tra * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Sha = (Sha * m.rac).sum(axis=-1).cumsum(axis=-1)
Om_Sra = (Sra * m.rac).sum(axis=-1).cumsum(axis=-1)

Om_tot = Om_Ts + Om_Tr + Om_Th + Om_Ss + Om_Sr + Om_Sh
```
