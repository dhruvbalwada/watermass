```python
from mitgcmdata import MITgcmmodel
import MITgcmutils.mds as mds
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.colors import SymLogNorm
import jmd95
%matplotlib inline
plt.rcParams['image.origin'] = 'bottom'
plt.rcParams['figure.figsize'] = (14,8)
```

```python
ddir = '/Users/rpa/RND/GCM/global_oce_latlon_layers/run'
m = MITgcmmodel.ModelInstance(ddir, default_iter=36001)
```

```python
T = m.rdmds('T', useMask=False)
S = m.rdmds('S', useMask=False)
rho, drhodt, drhods = np.zeros_like(T), np.zeros_like(T), np.zeros_like(T)
pref = 0.
for k in range(m.Nr):
    rho[k], drhodt[k], drhods[k]  = jmd95.eos.state(pref, T[k], S[k])
```

```python
# this is what is in my diagnostic file
# 'TOTTTEND' 'ADVr_TH ' 'ADVx_TH ' 'ADVy_TH ' 'DFrE_TH ' 'DFxE_TH ' 'DFyE_TH ' 'DFrI_TH '
d = m.rdmds('DiagGAD-T', useMask=False)

tottend = d[0]

ADVr_TH, ADVx_TH, ADVy_TH = d[1], d[2], d[3]
DFrE_TH, DFxE_TH, DFyE_TH, DFrI_TH = d[4],d[5],d[6],d[7]

# the layers version of the vertical diffusive flux 
dfx = m.rdmds('layers_dfx.01') 
dfy = m.rdmds('layers_dfy.01') 
dfr = m.rdmds('layers_dfr.01') 
afx = m.rdmds('layers_afx.01') 
afy = m.rdmds('layers_afy.01') 
afr = m.rdmds('layers_afr.01') 
```

```python
ld = m.rdmds('DiagLAYERS-diapycnal')
LaTs, LaTh, LaTr, LaTha, LaTra = ld[:5]
LaSs, LaSh, LaSr, LaSha, LaSra = ld[5:10]
LaHc, LaPic = ld[10:]

lb = mds.rdmds(m.output_dir + '/layers1RHO')
```

```python
vol = np.ma.masked_equal( m.hFacC * m.rac * m.drf, 0)
tend_adv = -vol**-1*( (ADVr_TH - np.vstack([ADVr_TH[1:], np.zeros((1,m.Ny,m.Nx))]))
            + (np.roll(ADVx_TH,-1,axis=2) - ADVx_TH)
            + (np.roll(ADVy_TH,-1,axis=1) - ADVy_TH))
tend_diff_rE = -vol**-1*( (DFrE_TH - np.vstack([DFrE_TH[1:], np.zeros((1,m.Ny,m.Nx))])) )
tend_diff_rI = -vol**-1*( (DFrI_TH - np.vstack([DFrI_TH[1:], np.zeros((1,m.Ny,m.Nx))])) )
tend_diff_xE = -vol**-1*( (np.roll(DFxE_TH,-1,axis=2) - DFxE_TH) )
tend_diff_yE = -vol**-1*( (np.roll(DFyE_TH,-1,axis=1) - DFyE_TH) )

tend_diff = tend_diff_xE + tend_diff_yE + tend_diff_rE + tend_diff_rI

tend_adv_x = -vol**-1*( (np.roll(ADVx_TH,-1,axis=2) - ADVx_TH) )
tend_adv_y = -vol**-1*( (np.roll(ADVy_TH,-1,axis=1) - ADVy_TH) )
tend_adv_r = -vol**-1*( (ADVr_TH - np.vstack([ADVr_TH[1:], np.zeros((1,m.Ny,m.Nx))])) )
tend_adv_h = tend_adv_x + tend_adv_y
tend_adv = tend_adv_h + tend_adv_r

tottend = np.ma.masked_array(tottend, vol.mask)
```

```python
# tendencies

i, j = 40, 16 # this works!
#i, j = 45, 10 # this is broken!!!
rf = m.rf.squeeze()[:-1]
rfp1 = m.rf.squeeze()[1:]
rc = m.rc.squeeze()

layers_adv = (afx+afy+afr)
layers_diff = (dfx+dfy+dfr)

plt.subplot(211)
plt.plot(tend_adv[:,j,i], rc, 'o-', tend_diff[:,j,i], rc, 'o-' )
plt.plot(layers_adv[:,j,i], rc, '.-', layers_diff[:,j,i], rc, '.-' )
plt.ylim([-500,0])
plt.legend(['adv', 'diff','adv (la)', 'diff (la)'])
plt.grid()

plt.subplot(212)
plt.plot(tend_adv[:,j,i], rc, 'o-', tend_diff[:,j,i], rc, 'o-' )
plt.plot(layers_adv[:,j,i], rc, '.-', layers_diff[:,j,i], rc, '.-' )
plt.ylim([-5000,0])
plt.xlim([-1e-8, 1e-8])
plt.grid()
```

```python
# pick a test point at which to plot the downward diffusive flux
i, j = 40, 15 # this works!
#i, j = 45, 10 # this is broken!!!
rf = m.rf.squeeze()[:-1]
rfp1 = m.rf.squeeze()[1:]
rc = m.rc.squeeze()
dfre = DFrE_TH[:,j,i] / m.rac[:,j,i]
dfri = DFrI_TH[:,j,i] / m.rac[:,j,i]
dftot = dfre + dfri
dfla = np.cumsum(dfr[:,j,i] * m.drf.squeeze() * m.hFacC[:,j,i])

aftot = ADVr_TH[:,j,i] / m.rac[:,j,i]
afrla = np.cumsum(afr[:,j,i] * m.drf.squeeze() * m.hFacC[:,j,i])


plt.subplot(211)
plt.plot(dftot, rf, 'o-', dfla, rfp1, '.:' )
plt.plot(aftot, rf, 'o-', afrla, rfp1, '.:' )
plt.ylim([-500,0])
plt.xlim([-5e-5, 5e-5])
plt.legend(['dftot', 'dfla', 'aftot', 'afla'])
plt.grid()

plt.subplot(212)
plt.plot(dftot, rf, 'o-', dfla, rfp1, '.:' )
plt.plot(aftot, rf, 'o-', afrla, rfp1, '.:' )
plt.ylim([-6000,0])
plt.xlim([-2e-6, 2e-6])
plt.grid()
```
